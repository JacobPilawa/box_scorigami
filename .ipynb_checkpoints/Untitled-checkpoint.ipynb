```python
import requests
import pandas as pd
import tqdm


import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.ticker as mticker
```

```python
# Example: get all games for a date range
url = "https://statsapi.mlb.com/api/v1/schedule"

params = {
    "sportId": 1,
    "startDate": "2023-03-01",
    "endDate": "2023-10-01"
}

data = requests.get(url, params=params).json()

games = []

for date in tqdm.tqdm(data["dates"]):
    for game in date["games"]:
        game_id = game["gamePk"]

        # Fetch boxscore
        box_url = f"https://statsapi.mlb.com/api/v1/game/{game_id}/boxscore"
        box = requests.get(box_url).json()

        teams = box["teams"]

        for side in ["away", "home"]:
            stats = teams[side]["teamStats"]["batting"]
            games.append({
                "game_id": game_id,
                "team": teams[side]["team"]["name"],
                "runs": stats["runs"],
                "hits": stats["hits"],
                "errors": teams[side]["teamStats"]["fielding"]["errors"]
            })

df = pd.DataFrame(games)
print(df.head())
```

```python
df.to_parquet('scrape.parquet')
```

```python
# Global max count across all panels
counts = df.groupby(['hits', 'runs', 'errors']).size()
global_max = int(counts.max()) if len(counts) > 0 else 1

error_values = sorted(df['errors'].unique())
n_panels = len(error_values)
n_cols = 3
n_rows = (n_panels + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows), dpi=250)
axes = axes.flatten()

max_hits = df['hits'].max()
max_runs = df['runs'].max()

cmap = plt.cm.copper.copy()
cmap.set_under('white')

for i, err in enumerate(error_values):
    subset = df[df['errors'] == err]

    heatmap, xedges, yedges = np.histogram2d(
        subset['hits'], subset['runs'],
        bins=[range(0, max_hits + 2), range(0, max_runs + 2)]
    )

    heatmap = np.ma.masked_where(heatmap == 0, heatmap)
    im = axes[i].imshow(
        heatmap.T,
        origin='lower',
        aspect='auto',
        cmap=cmap,
        interpolation='nearest',
        #vmin=1,
        #vmax=global_max,
        norm=LogNorm(vmin=1, vmax=100)
    )

    axes[i].set_title(f'Errors = {err}', fontsize=15)
    axes[i].set_xlabel('Hits', fontsize=15)
    axes[i].set_ylabel('Runs', fontsize=15)

    # Major ticks every integer
    major_ticks_x = np.arange(0, max_hits + 1, 1)
    major_ticks_y = np.arange(0, max_runs + 1, 1)
    axes[i].set_xticks(major_ticks_x)
    axes[i].set_yticks(major_ticks_y)

    # Label only every 5
    axes[i].set_xticklabels(
        [str(t) if t % 5 == 0 else '' for t in major_ticks_x],
        fontsize=10
    )
    axes[i].set_yticklabels(
        [str(t) if t % 5 == 0 else '' for t in major_ticks_y],
        fontsize=10
    )

    # Minor ticks for grid alignment
    axes[i].set_xticks(np.arange(-0.5, max_hits + 1, 1), minor=True)
    axes[i].set_yticks(np.arange(-0.5, max_runs + 1, 1), minor=True)
    axes[i].tick_params(which='minor', length=0)
    axes[i].grid(which='minor', color='gray', linestyle='-', linewidth=0.1)

    # Colorbar with global scale
    divider = make_axes_locatable(axes[i])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    cb = plt.colorbar(im, cax=cax, label='$N_{{{results}}}$')

    cb.locator = LogLocator(base=10)
    cb.formatter = ScalarFormatter()   # <-- key change
    cb.formatter.set_scientific(False) # force no scientific notation
    cb.formatter.set_useOffset(False)
    
    cb.update_ticks()
    ticks = cb.get_ticks()[1:-1]
    cb.set_ticks(ticks)
    cb.set_ticklabels([f"{int(t)}" for t in ticks])


# Turn off unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

for ax in axes:
    ax.set_aspect('equal')

plt.suptitle('MLB Game Outcomes by Hits, Runs, and Errors (2023 Season)', fontsize=25, y=0.95)
plt.tight_layout()
plt.subplots_adjust(hspace=-0.1, wspace=0.3)
plt.savefig('box_scorigami_2015_2026.png',dpi=250,bbox_inches='tight')
plt.show()
```
